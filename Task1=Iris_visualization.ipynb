```python
# Task 1: Exploring and Visualizing the Iris Dataset

This notebook explores the famous Iris flower dataset to understand the structure of the data and visualize it using scatter plots, histograms, and box plots. These visualizations help reveal patterns between different flower species based on petal and sepal measurements.
```

```python
# Importing Libraries and Loading the Data below now
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Show the first few rows of the dataset
iris.head()
```

```python
## Dataset Description

The Iris dataset contains 150 rows and 5 columns. Each row represents a different flower sample, with the following features:

- **sepal_length**: Length of the sepal in cm
- **sepal_width**: Width of the sepal in cm
- **petal_length**: Length of the petal in cm
- **petal_width**: Width of the petal in cm
- **species**: Type of iris flower (Setosa, Versicolor, or Virginica)
```

```python
# Lets Explore the Dataset Structure now.
```

```python
# Shape of the dataset
print("Shape of dataset:", iris.shape)
```

```python
# Column names
print("Column names:", iris.columns)
```

```python
# Dataset information
iris.info()
```

```python
## Scatter Plot: Petal Length vs Petal Width

This plot helps visualize how the petal size varies between different iris species.
```

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
#Lets work on Histogram now.
## Histogram: Sepal Length Distribution
This shows the distribution (spread) of sepal lengths across all flowers.
```

```python
sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
## Box Plot: Sepal Width by Species
```

```python
sns.boxplot(data=iris, x="species", y="sepal_width")
plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
## Conclusion

#The Iris dataset shows clear differences between species in terms of petal length and width.
```

```python
#The scatter plot demonstrates strong separation between species, especially for Setosa. Histograms help in understanding the spread of feature values, while box plots show outliers and variation.
```
