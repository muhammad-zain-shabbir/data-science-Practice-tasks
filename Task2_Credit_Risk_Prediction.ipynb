The goal of this task is to predict whether a loan applicant is likely to default or not, based on details like gender, income, loan amount, credit history, and more.


```python
import pandas as pd

# Load the correct file using the exact name
df = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")

# Show the first few rows to confirm it's working
df.head()
```

```python
## Dataset Understanding

We are using a dataset from a past loan prediction challenge. It includes features such as:

- Gender, Married, Dependents
- Education, Self Employed
- Applicant Income, Coapplicant Income
- Loan Amount and Loan Term
- Credit History
- Property Area
- Loan Status (the target variable: Y or N)
```

## Data Cleaning and Preparation

We handled missing values using the most common (mode) or median values. This ensures that our data is complete and ready for analysis and modeling.


```python
df.shape
df.columns
df.info()
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
```

```python
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
                        
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```
